1. Each variable forms a column.
2. Each observation forms a row.
3. Each type of observational unit forms a table.

### Analysis

All columns after the religion column are values of a variable. Thus, according to the rules of tidy data, it should be grouped under one column name.

```python
# Normally, the value_vars parameter should be defined in the melt(), such as, value_vars=['<$10k', '$10-20k',...].
# Since all of the columns after the religion should be included value_vars. A trick is to only defined the id_vars
# in melt(), everything isn't included in the id_vars will be added to value_vars by pandas automatically.
df.melt(id_vars='religion',
        var_name='income',
        value_name='freq',).sort_values(by='religion').head(10)
```

### Analysis

1. All the columns labeled as x1st.week...x76th.week are values of one variable. Thus, they should be grouped under one column name
2. Each week's value should be represented by 1,2,3...76 rather x1st.week, x2nd.week, etc

```python
df_melt = df.melt(id_vars=['year', 'artist.inverted', 'track', 'time', 'genre', 'date.entered','date.peaked'],
                  var_name = 'week',
                  value_name = 'rank'
                 )
df_melt.head()
```

```python
df_melt.dtypes
```

After melting the data, notice the data type for both `week` and `rank` is object (which is stored as string). They should be convereted to the int type.

```python
# when there is only one capture group and `expand=False`, `extract` returns a Series rather than a DataFrame
df_melt.week = df_melt.week.str.extract('(\d+)', expand=False).astype('int')
```

```python
df_melt['rank'].astype(int)
```

The error is caused by the `NaN` values in the `rank` column.

```python
df_melt['rank'].value_counts(dropna=False)
```

If a song is dropped out of top 100, its rank value is `NaN` in the original dataset. Notice after week 7, the rank value is all `NaN`. These values need to be dropped from the dataset.

```python
df_melt.sort_values(by=["year","artist.inverted","track","week","rank"]).head(15)
```

```python
df_melt = df_melt.dropna()
df_melt.sort_values(by=["year","artist.inverted","track","week","rank"]).head(10)
```

```python
df_melt['rank'] = df_melt['rank'].astype('int')
```

### Tubercolosis Example

A few notes on the raw data set:

The columns starting with "m" or "f" contain multiple variables:
Sex ("m" or "f")
Age Group ("0-14","15-24", "25-34", "45-54", "55-64", "65", "unknown")
Mixture of 0s and missing values("NaN"). This is due to the data collection process and the distinction is important for this dataset.

```python
df = pd.read_csv("../data/tb-raw.csv")
df
```

```python
df = pd.melt(df, id_vars=["country","year"], value_name="cases", var_name="sex_and_age")

# Extract Sex, Age lower bound and Age upper bound group
tmp_df = df["sex_and_age"].str.extract("(\D)(\d+)(\d{2})", expand=False)    

# Name columns
tmp_df.columns = ["sex", "age_lower", "age_upper"]

# Create `age`column based on `age_lower` and `age_upper`
tmp_df["age"] = tmp_df["age_lower"] + "-" + tmp_df["age_upper"]

# Merge 
df = pd.concat([df, tmp_df], axis=1)

# Drop unnecessary columns and rows
df = df.drop(['sex_and_age',"age_lower","age_upper"], axis=1)
df = df.dropna()
df = df.sort_values(ascending=True,by=["country", "year", "sex", "age"])
df.head(10)
```

### Global Historical Climatology Network Dataset¶

```python
df = pd.read_csv("../data/weather-raw.csv")
df = pd.melt(df, id_vars=["id", "year","month","element"], var_name="day_raw")
df.head(10)
```

```python
import datetime
# Extracting day
df["day"] = df["day_raw"].str.extract("d(\d+)", expand=False)  
df["id"] = "MX17004"

# To numeric values
df[["year","month","day"]] = df[["year","month","day"]].apply(lambda x: pd.to_numeric(x, errors='ignore'))

# Creating a date from the different columns
def create_date_from_year_month_day(row):
    return datetime.datetime(year=row["year"], month=int(row["month"]), day=row["day"])

df["date"] = df.apply(lambda row: create_date_from_year_month_day(row), axis=1)
df = df.drop(['year',"month","day", "day_raw"], axis=1)
df = df.dropna()

# Unmelting column "element"
df = df.pivot_table(index=["id","date"], columns="element", values="value")
df.reset_index(drop=False, inplace=True)
df
```

### One type in multiple tables

```python
import glob
import re

def extract_year(string):
    match = re.match(".+(\d{4})", string) 
    if match != None: return match.group(1)
    
path = '../data'
allFiles = glob.glob(path + "/201*-baby-names-illinois.csv")
frame = pd.DataFrame()
df_list= []
for file_ in allFiles:
    df = pd.read_csv(file_,index_col=None, header=0)
    df.columns = map(str.lower, df.columns)
    df["year"] = extract_year(file_)
    df_list.append(df)
    
df = pd.concat(df_list)
df.head(5)
```
